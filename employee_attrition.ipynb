## Using pickle to download method to a file

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("employee_data.csv")
```

```python
df.head(5)
```

```python
y=df['Attrition'].map({'Yes':1, 'No':0})
x=df[['Age', 'NumCompaniesWorked', 'JobSatisfaction' , 'PerformanceRating' ,'MonthlyIncome', 'JobLevel' , 'DistanceFromHome']]
```

```python
 #Standard Scalar for Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5, random_state=0)
```

```python
#Building logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=50)
model.fit(X_train,y_train)
y_predicted = model.predict(X_test)
```

```python
import pickle
```

```python
pickle.dump(model, open('model.pkl', 'wb'))
```

## To load a model from a pickle file

```python
pickled_model = pickle.load(open('model.pkl', 'rb'))
pickled_model.predict(X_test)
```
